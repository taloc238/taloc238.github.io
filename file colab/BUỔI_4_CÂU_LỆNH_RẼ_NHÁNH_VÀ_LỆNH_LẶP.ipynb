**Bài tập 5.1**: ***Chia táo***: Thuý Kiều là chị, em là Thuý Vân. Hôm nay, bà đi chợ. Vì bà biết hai
cháu của mình rất thích ăn táo nên bà đã mua quà cho Kiều và Vân là 1kg táo. Bà bảo rằng: Kiều hãy
chia đều số táo bà mua cho hai chị em nếu lẻ ra một quả Kiều phải cho em quả đó để em được phần
hơn. Hãy viết chương trình để chia táo giúp Kiều.
+ Ràng buộc: Số quả táo là số nguyên dương được nhập từ bàn phím.
+ Đầu vào: Nhập vào số quả táo bà mua cho Thuý Kiều và Thuý Vân.
+ Đầu ra: Số lượng táo của Thuý Kiều, Thuý Vân nhận được là số nguyên .

```python
quatao = int(input("Nhập vào số quả táo bà mua: "))
tao_Kieu = quatao//2
tao_Van = quatao - tao_Kieu
if quatao%2 ==0:
    print("Số táo Kiều nhận được là:",int(tao_Kieu))
    print("Số táo Vân nhận được là:",int(tao_Van))
else:
    print("Số táo kiều nhận được là:",int(tao_Kieu))
    print("Số táo Vân nhận được là:",int(tao_Van))
```

1. Lặp có số lần lặp biết trước và cấu trúc lặp .

Cú pháp:

Cú pháp: <biến đếm> *in* <dữ liệu>:

Hàm range()

Cú pháp: range (start, stop(end), step)

start: biến bắt đầu

stop: biến kết thúc (giá trị = stop - 1)

step: bước nhảy (tùy biến nhưng mặc định 1 đơn vị số nguyên)

```python
i = 3
for j in range (1,10,1):
    print(i,"*",j, "=",i*j)
```

```python
i = 3
for j in range (1,10,2):
    print(i,"*",j, "=",i*j)
```
