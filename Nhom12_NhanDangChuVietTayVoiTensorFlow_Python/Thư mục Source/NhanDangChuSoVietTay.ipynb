```python
import os
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Quyết định huấn luyện mô hình mới hay dùng mô hình có sẵn
train_new_model = True

if train_new_model:
    # Tải và xử lý dữ liệu Mnist
    mnist = tf.keras.datasets.mnist
    (X_train, y_train), (X_test, y_test) = mnist.load_data()

    # Chuẩn hóa dữ liệu
    X_train = X_train / 255.0
    X_test = X_test / 255.0

    # Xây dựng mô hình dạng nơ-ron
    model = tf.keras.models.Sequential([
        tf.keras.layers.Input(shape=(28, 28)),          
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(10, activation='softmax')
    ])

    # Biên dịch mô hình
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    # Huấn luyện mô hình
    model.fit(X_train, y_train, epochs=15, validation_data=(X_test, y_test))

    # Đánh giá mô hình
    val_loss, val_acc = model.evaluate(X_test, y_test)
    print("Validation loss:", val_loss)
    print("Validation accuracy:", val_acc)

    # Lưu mô hình đã huấn luyện
    model.save('handwritten_digits.keras')
else:
    # Tải mô hình đã có
    model = tf.keras.models.load_model('handwritten_digits.keras')

# Dự đoán số từ ảnh bên ngoài
image_number = 1
while os.path.isfile(f'digits/digit{image_number}.png'):
    # Xử lý ảnh từ bên ngoài
    try:
        img = cv2.imread(f'digits/digit{image_number}.png', cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (28, 28))  
        img = np.invert(img) 
        img = img / 255.0  
        img = img.reshape(1, 28, 28)
    
    # Dự đoán số trong ảnh
        prediction = model.predict(img)
        print(f"The number is probably a {np.argmax(prediction)}")
    
        plt.imshow(img[0], cmap=plt.cm.binary)
        plt.show()
    
    # xử lý lỗi
        image_number += 1
    except Exception as e:
        print(f"Error reading image {image_number}: {e}")
        image_number += 1

```
